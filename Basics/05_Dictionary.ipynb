```python
#Dictionary {}
#A dictionary stores the data in key-value pairs
```

```python
Student={
    "Name" : "Tarun",
    "Age " : 30,
    "Course" :"Python",
    4:"India"
}
print(Student["Name"])   #Extrcating the value by using the key
print(Student[4])
```

```python
#how to add the new pair
Student["company"]="ABC"
print(Student)

#How to update the value
Student["Course"]="Cloud"
print(Student)
```

```python
#Dictionary methods
#keys---will give only the keys in the dict
print(Student.keys())

#values---will give only the values in the dict
print(Student.values())

#items---it will give keys and values as the pair
print(Student.items())


print(Student)
```
