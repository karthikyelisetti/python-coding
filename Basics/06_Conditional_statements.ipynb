```python
#Student grade checker
#we want to assign grades based on the marks

#90 and above ---->Grade A
#75 to 89 ---->Grade B
#50 to 74 ---->Grade C
#below 50 -->fail (else)


marks =120

if marks>=90:
  print("Grade A")
elif marks>=75:
  print("Grade B")
elif marks >=50:
  print("Grade C")
else:
  print("Fail")
```
