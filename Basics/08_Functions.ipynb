```python
#Create a function check_postive_negative
#Check if a number is positive or negative
#take user input
#it should handle zero also (if, elif, else)

def check_positive_negative():
    n = int(input("Enter the number: "))

    if n == 0:
        print("Number is zero")
    elif n > 0:
        print("Positive number")
    else:
        print("Negative number")

check_positive_negative()
```

```python
# Create two empty lists with name odd and even
#Using for loop with range (1 to 20)
# and inside loop using conditional statements and you should add
# even number into even list and odd numbers into odd list

def even_odd_list():

    odd = []
    even = []

    for i in range(1, 21):
        if (i%2 == 0):
            even.append(i)
        else:
            odd.append(i)

    print(even)
    print(odd)

even_odd_list()
```

```python
#By using while loop (single loop)
#Output should be

# *
# **
# ***
# ****
# *****

def pattern():
    i = 1

    while (i <= 5):
        print(i * "*")
        i=i+1

pattern()
```

```python
#text = "banana"
#count how many letter 'a' are there in the string

str = input("Enter the string")
l = input("input the letter to find")

def countLetter(str, l):
    count = 0

    for i in str:
        if i == l:
            count += 1
    
    print(count)

countLetter(str, l)

```

```python
#create a function which takes a single parameter
#if the number is divisible by 2 or 3
#then should print divisible by 2 or 3
#else should not divisible bt 2 or 3

n = int(input("Enter the number"))

def divisible(n):
    if ((n % 2 == 0) or (n % 3 == 0)):
        print("Number is divisible by 2 or 3")
    else:
        print("Number is not divisible by 2 or 3")

divisible(n)
```

```python
#text = "Python"
# and count how many vowels are there in the text

text = "Python"

def countVowels(text):
    count = 0
    vowels = "aeiou"

    for i in text:
        if i in vowels:
            count = count + 1

    print (count)

countVowels(text)
```
