```python
#what is identity?
#identity operator will check whether two variables are pointing to the sameobject or not.
#is
#is not

a=[1,2,3]

b=a

c=[1,2,3]

print(a is c)    #false
print(a is b)   #True


#Membership operator
#it will check whether the value is exists in the collection or not
# in
# not in

nums=[10,20,30]

print(20 in nums)  #True
print(50 in nums)  #False
```
