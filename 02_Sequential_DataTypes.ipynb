```python
#list
#we can create list using []
```

```python
a=[10,20,30,"Student","India",89.8]
print(type(a))
print(a[4])
print(a[5])

print(a[-3])
print(a[-1])
print(a[5])
```

```python
#Slicing---> when u want to extract any specific part from whole we will be using slicing.
#Syntax -->variablename[startindex :endindex+1]

#syntax with step ---->variablename[startindex : endindex+1 :stepcount+1]
```

```python
lst = [10,20,30,40,50,60,70]

#if i want from 20 - 50
print(lst[1:5])

#10-40
print(lst[0:4])

#30-50
print(lst[2:5])

#20-60
print(lst[1:6])
```

```python
a= [10 , 20 , "Tarun" , "sony" , "Nishkarsh" , 50 , 80.4]

#output --->20,sony,50
#20-50
print(a[1:6:2])

# i want output as 10 and tarun
print(a[0:3:2])
```

```python
b = ["Tarun" , "Sony" , "Nishkarsh" , "Dipansh" , "Raghavendra" , "Dilya"]

#from tarun to raghavendra
#i want only tarun,nishkarsh,raghavendra

print(b[0:5:2])
```

```python
#Methods in the list

#append() ---> will add the elemenet to the end of the list
b.append("Karthik")
print(b)
```

```python
#insert() --> will add the elment in the specific position
b.insert(2,"Saran")
print(b)
```

```python
b.insert(0,"Kapil")
print(b)
```

```python
b[4] = "Varun"
print(b)
```

```python
#pop() --->will remove the elements.(by default it will remove the last element)
b.pop()
print(b)
```

```python
#pop(index number)
b.pop(0)
print(b)
```

```python
#Tuple methods

tup = (20 , "Tarun" , "Sony" , "Nishkarsh" , "Karthik" , 89.0 , "Tarun")
print(tup)

#count() ----> used to count the number of occurences of the element
print(tup.count("Tarun"))

#index() ---> it will be giving the index nuber of the element
print(tup.index(89.0))
```
